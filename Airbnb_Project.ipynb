```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("airbnb_raw_data.csv")
```

```python
df.head()
```

```python
df.columns
```

# Checking Missing Values

```python
print (df.isnull().sum())
```

# Handling Missing Values

```python
df ['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month': 0, 'last review': df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=["license", "house_rules"], errors = 'ignore')
```

```python
df.head()
```

```python
# Remove Dollar Signs and Convert to Float
df['price'] = df['price'].replace('[\$,]', '' , regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '' , regex=True).astype(float)
```

```python
df.head()
```

# Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

# Descriptive Statistics

```python
df.describe()
```

# Visualization

# What is the distribution of listing price?

```python
plt.figure(figsize = (10,6) )
sns.histplot(df['price'], bins = 50, kde = True, color = 'black')
plt.title("DISTRIBUTION OF LISTING PRICE")
plt.xlabel("PRICE $")
plt.ylabel("FREQUENCY")
plt.show()
```

```python
The histogram shows a fairly even distribution of 
listing prices across different price ranges, indicating no 
particular concentration of listings in any specific price range. 
The KDE line helps visualize this even spread more clearly, 
confirming that the dataset contains listings with a wide varietyÂ ofÂ prices.
```

# How are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize = (10,6) )
sns.countplot(x = "room type", data = df, color = 'black')
plt.title("ROOM TYPE DISTRIBUTION")
plt.xlabel("ROOM TYPE")
plt.ylabel("COUNT")
plt.show()
```

# How are listings distributed across different neighbourhoods?

```python
plt.figure(figsize = (10,6) )
sns.countplot(y = 'neighbourhood group', data = df, color = 'black', order = df['neighbourhood group'].value_counts().index)
plt.title("NUMBER OF LISTINGS BY NEIGHBOURHOOD GROUP")
plt.xlabel("COUNT")
plt.ylabel("NEIGHBOURHOOD GROUP")
plt.show()
```

# What is the relationship between price and room type?

```python
plt.figure(figsize = (10,6) )
sns.boxplot(x ='room type', y ='price', hue ='room type', data = df, palette='Set2')
plt.title('PRICE VS ROOM TYPE')
plt.xlabel('ROOM TYPE')
plt.ylabel('PRICE ($)')
plt.legend(title='Room Type')
plt.show()
```

# How has the number of reviews change over with time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize = (12,6) )
reviews_over_time.plot(kind = 'line', color = 'red')
plt.title('NUMBER OF REVIEWS OVER TIME')
plt.xlabel('DATE')
plt.ylabel('NUMBER OF REVIEWS')
plt.show()
```
